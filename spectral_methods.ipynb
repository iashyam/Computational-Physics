```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import fft
from numpy import pi

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
def f1(x):
    return np.sin(x)
```

### Diffrentiation using fft

The differentiation operator directly translate into simple multiplication in the fourier space.

say $$f'(x) = \frac{df}{dx}$$.

If we take the fourier transform both side this will become

$$\hat{f'}(x) = \iota k \hat{f}$$

Now we can take the inverse trasform to calculate the diffrentition.

$$f'(x) = \mathcal{F}^{-1} (\iota k \hat{f})$$

```python
N = 128
xxs = np.linspace(-2*pi, 2*pi, N)
yys = f1(xxs)

h = xxs[1] - xxs[0]

plt.plot(xxs, yys)
```

```python
y_hat = fft.fft(yys)
freqs = fft.fftfreq(len(xxs))*(2*pi/h)

#multiplying by fourer difere
diff_y_hat = 1j*freqs*y_hat

#taking inverse fourier transform
diff_y_hat_real = np.real(fft.ifft(diff_y_hat))

plt.plot(xxs, diff_y_hat_real, label='using fft')
plt.plot(xxs, np.cos(xxs), label = 'analytical')
plt.legend()
plt.show()
```

# integrating a function using fft

The differentiation operator directly translate into simple multiplication in the fourier space.

say $$f'(x) = \int f(x)  dx$$.

If we take the fourier transform both side this will become

$$\hat{f'}(x) = \frac{\hat{f}}{\iota k}$$

Now we can take the inverse trasform to calculate the diffrentition.

$$f'(x) = \mathcal{F}^{-1} (\frac{\hat{f}}{\iota k})$$

```python
y_hat = fft.fft(yys)
freqs = fft.fftfreq(len(xxs))*(2*pi/h)

#multiplying by fourer difere
diff_y_hat = y_hat/1j*freqs

#taking inverse fourier transform
diff_y_hat_real = np.real(fft.ifft(diff_y_hat))

plt.plot(xxs, diff_y_hat_real,label='fft')
plt.plot(xxs, -np.cos(xxs), label='real')
plt.legend()
plt.show()
```
