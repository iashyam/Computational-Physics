## 2-D Matplotlib plot

Matplotlib provides a goot amount of visualisation tools:
#### Simple Function Plot

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
xs = np.linspace(-2*math.pi, 2*math.pi, 1000)
ys = -np.sin(2*xs)*np.cos(3*xs)
```

```python
plt.plot(xs,ys, '-r', label='f(x)')
plt.xlabel(' xs ')
plt.ylabel('ys')
plt.title('F(x) vs x')
# plt.grid(True)
plt.legend()
```

#### Matplotlib Subplots

```python
def f1(x):
    return np.sin(x)

def f2(x):
    return np.sin(2*x)*np.cos(x)

def f3(x):
    return np.sin(x)*np.cos(2*x)
    
def f4(x):
    return np.cos(x)
    
fig, ax = plt.subplots(2,2)

#feed the values in list ax
ax[0][0].plot(xs, f1(xs))
ax[0][1].plot(xs, f2(xs))
ax[1][0].plot(xs, f3(xs))
ax[1][1].plot(xs, f4(xs))
```
